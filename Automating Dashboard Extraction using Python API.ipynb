```python
from pptx import Presentation 
from pptx.util import Inches 
import glob
def create_ppt(item):
# Creating an Presentation object
    i=0
    ppt = Presentation(os.path.abspath('./Q2 2021 Template.pptx')) 

    # Selecting blank slide
    blank_slide_layout = ppt.slide_layouts[5] 

    #ppt.slide_height = 5143500
    for g in sorted(glob.glob("./Talent_Dashboard/data_raw/*"),key=os.path.getmtime):
        # Giving Image path 
        img_path = g
        i=i+1
        # Attaching slide to ppt
        #slide = ppt.slides.add_slide(blank_slide_layout) 
        slide=ppt.slides[i]
        if 'Overview' in img_path or 'Acquire Great talent-1' in img_path or 'Retain Great Talent' in img_path:
        # For margins
        #left = (ppt.slide_width*0.15)
            left=Inches(0.5)
            top = Inches(0.41)
            height= Inches(6.15)
            #right=0.1
            width = Inches(9.1)
        elif 'Performance-Engagement-2' in img_path:
            left=Inches(0.78)
            top = Inches(0.25)
            height= Inches(6.32)
            #right=0.1
            width = Inches(6.46)
            
        else:
            left=Inches(0.5)
            top = Inches(1.16)
            height= Inches(4.65)
            #right=0.1
            width = Inches(9.2)
            
            

        # adding images
        #pic = slide.shapes.add_picture(img_path,
                                       #left, top)
            
        pic = slide.shapes.add_picture(img_path, left, top, width, height)


        
    # Save file
    ppt.save(f'./Talent_Dashboard/ppt/{item}.pptx')

    print("Done")  
```

```python
def capture(list_filter,dashboard_view_id_list,Tab_Names,filter_val):
    flag=0
    for item in list_filter:

        for ele,name in zip(dashboard_view_id_list,Tab_Names):
            dashboard_view_id=ele

            pdf_params = {
            'type': 'type=A4',
            'orientation': 'orientation=Landscape',
            'filter': None }

    #Generate images #Filering the type of filter application over the Tableau Server
            if filter_val=='Region': 
                pdf_params['filter'] = f'vf_Region={item}'
            elif filter_val=='Exec':
                pdf_params['filter'] = f'vf_P4={item}'
            elif filter_val=='Country':
                pdf_params['filter'] = f'vf_Country={item}'
            elif filter_val=='Exec-Country':
                flag=1
                pdf_params['filter'] = f'vf_P4={item[0]}&vf_Country={item[1]}'
            elif filter_val=='Exec-Region':
                flag=1
                pdf_params['filter'] = f'vf_P4={item[0]}&vf_Region={item[1]}'
            elif filter_val=='Country-Country':
                flag=1
                pdf_params['filter'] = f'vf_Country={item[0]},{item[1]}'
            elif filter_val=='Region-Dep':
                flag=1
                pdf_params['filter'] = f'vf_Region={item[0]}&vf_Department={item[1]}'
            elif filter_val=='Region-Dep-2':
                flag=2
                pdf_params['filter'] = f'vf_Region={item[0]},{item[1]}&vf_Department={item[2]}'
            
                
             
            img = conn.query_view_image(view_id=dashboard_view_id, parameter_dict=pdf_params)
            
            if flag==0:
                with open(os.path.join(os.path.abspath('./Talent_Dashboard/data_raw'),f'{item}-{name}.png'), 'wb') as img_file:
                    img_file.write(img.content)
                crop_overview(item,name)
            elif flag==1:
                with open(os.path.join(os.path.abspath('./Talent_Dashboard/data_raw'),f'{item[0]}-{item[1]}-{name}.png'), 'wb') as img_file:
                    img_file.write(img.content)
                crop_overview(item[0]+'-'+item[1],name)
            elif flag==2:
                with open(os.path.join(os.path.abspath('./Talent_Dashboard/data_raw'),f'{item[0]}-{item[1]}-{item[2]}-{name}.png'), 'wb') as img_file:
                    img_file.write(img.content)
                crop_overview(item[0]+'-'+item[1]+'-'+item[2],name)
                
                
            
            
        if flag==0:
        #Calling the ppt generate function
            create_ppt(item)

            #Cleaning up the raw_data
            remove_files()
        elif flag==1:
            create_ppt(item[0]+'-'+item[1])

            #Cleaning up the raw_data
            remove_files()
        elif flag==2:
            create_ppt(item[0]+'-'+item[1]+'-'+item[2])

            #Cleaning up the raw_data
            remove_files()
            
            
```

```python
import os
def createfolder(newpath):
    if not os.path.exists(newpath):
        os.makedirs(newpath)
    else:
        remove_files()

createfolder('./Talent_Dashboard/data_raw/')
createfolder('./Talent_Dashboard/ppt/')
```
